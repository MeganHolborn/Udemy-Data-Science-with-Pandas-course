### Import data:

__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).

* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

### Sort and filter data to find the best and worst movies with the:
- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity


```python
# Sort by highest revenue
movies.sort_values(by=["revenue_musd"], ascending=False)[["title","revenue_musd"]]
```

```python
# Sort by highest budget
movies.sort_values(by=["budget_musd"], ascending=False)[["title","budget_musd"]]
```

```python
# Sort by highest profit (revenue - budget)
movies_profit = movies[["title", "revenue_musd", "budget_musd"]]
movies_profit["profit_musd"] = movies_profit["revenue_musd"] - movies_profit["budget_musd"]
movies_profit.sort_values(by=["profit_musd"], ascending=False)
```

```python
# Sort by lowest profit (revenue - budget)
movies_profit.sort_values(by=["profit_musd"], ascending=True)
```

```python
# Sort by highest return on investment
movies_roi = movies[["title", "revenue_musd", "budget_musd"]]
movies_roi["return_on_investment_musd"] = movies_roi["revenue_musd"] / movies_roi["budget_musd"]
movies_roi.sort_values(by=["return_on_investment_musd"], ascending=False)
```

```python
# Sort by lowest return on investment
movies_roi.sort_values(by=["return_on_investment_musd"], ascending=True)
```

```python
# Sort by highest number of votes
movies.sort_values(by=["vote_count"], ascending=False)[["title", "vote_count"]]
```

```python
# Sort by highest rating in movies with 10 or more ratings
# First filter by movies that have more than 10 votes, so that the rating data is reliable
movies[movies.vote_count>10].sort_values(by=["vote_average"], ascending=False)[["title","vote_count","vote_average"]]
```

```python
# Sort by lowest rating in movies with 10 or more ratings
# First filter by movies that have more than 10 votes, so that the rating data is reliable
movies[movies.vote_count>10].sort_values(by=["vote_average"], ascending=True)[["title","vote_count","vote_average"]]
```

```python
# Highest popularity
movies.sort_values(by=["popularity"], ascending=False)[["title","popularity"]]
```

```python
# Create a user defined function to find the n best or worst movies based on a minimum user vote and revenue, sort by a user defined column
def movie_filter(dataset, n, sort_by, ascending, min_vote_average, min_revenue):
    dataset = pd.DataFrame(dataset)
    filters = (dataset.vote_average > min_vote_average) & (dataset.revenue_musd > min_revenue)
    filtered_dataset = dataset[filters].sort_values(by=[sort_by], ascending=ascending).head(n)
    return filtered_dataset[["title", sort_by, "vote_average", "revenue_musd"]]

movie_filter(movies, 5, "popularity", False, 5, 2)
```

## Find your next Movie
3. __Filter__ the Dataset for movies that meet the following conditions:

__Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)__

__Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)__

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

__Search 4: Action or Thriller Movie with original language English and minimum Rating of 7.5 (most recent movies first)__

```python
# Filter dataset for science fiction, action movies with Bruce Willis (sorted from high to low rating)
genre_filter = movies.genres.str.contains("science fiction", False) & movies.genres.str.contains("action", False)
cast_filter = movies.cast.str.contains("Bruce Willis", False)

movies[genre_filter & cast_filter][["title", "genres", "cast", "vote_average"]].sort_values(by=["vote_average"], ascending=False)
```

```python
# Filter for movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)
cast_filter = movies.cast.str.contains("Uma Thurman", False)
director_filter = movies.director == ("Quentin Tarantino")

movies[cast_filter & director_filter][["title","cast","director","runtime"]].sort_values(by=["runtime"], ascending=True)
```

```python
# Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)
production_com_filter = movies.production_companies.str.contains("Pixar", False)
date_filter=movies.release_date.between("2010-01-01", "2015-12-31")
movies[production_com_filter & date_filter][["title","production_companies","release_date","revenue_musd"]].sort_values(by=["revenue_musd"], ascending=False)
```

```python
# Action or Thriller Movie with original language English and minimum Rating of 7.5 (most recent movies first)
genre_filter = movies.genres.str.contains("thriller", False) | movies.genres.str.contains("thriller", False)
language_filter = movies.original_language == ("en")
rating_filter = movies.vote_average >= 7.5

movies[genre_filter & language_filter & rating_filter][["title","genres","original_language","vote_average","release_date"]].sort_values(by=["release_date"],ascending=False)
```

## Are Franchises more successful?
4. __Analyze__ the Dataset and __find out whether Franchises (Movies that belong to a collection) are more successful than stand-alone movies__ in terms of:

- mean revenue
- median Return on Investment
- mean budget raised
- mean popularity
- mean rating


```python
# Create a subset will all data columns of interest and calculate ROI
movies["ROI"] = movies.revenue_musd / movies.budget_musd
movies_subset = movies[["title","belongs_to_collection","revenue_musd","budget_musd","popularity","vote_average", "ROI"]]
movies_subset
```

```python
# Create filters to select franchises or stand-alone movies
franchises = movies_subset.belongs_to_collection.notna()
stand_alones = movies_subset.belongs_to_collection.isna()
```

```python
# Compare franchise or stand-alone movie statistics
franchise_revenue = movies_subset.groupby(franchises).revenue_musd.mean()[1]
stand_alone_revenue = movies_subset.groupby(stand_alones).revenue_musd.mean()[1]
franchise_ROI = movies_subset.groupby(franchises).ROI.mean()[1]
stand_alone_ROI = movies_subset.groupby(stand_alones).ROI.mean()[1]
franchise_budget = movies_subset.groupby(franchises).budget_musd.mean()[1]
stand_alone_budget = movies_subset.groupby(stand_alones).budget_musd.mean()[1]
franchise_pop = movies_subset.groupby(franchises).popularity.mean()[1]
stand_alone_pop = movies_subset.groupby(stand_alones).popularity.mean()[1]
franchise_ratings = movies_subset.groupby(franchises).vote_average.mean()[1]
stand_alone_ratings = movies_subset.groupby(stand_alones).vote_average.mean()[1]
```
